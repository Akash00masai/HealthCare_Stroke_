```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# Load dataset
df = pd.read_csv('cleaned_healthcare_dataset_stroke.csv')
```

```python

# Set the aesthetic style of the plots
sns.set(style="whitegrid")
```

```python

# 1. Histogram: Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True, color='#1f77b4')  # Blue
plt.title('Age Distribution', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# 2. Histogram: Glucose Levels Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['avg_glucose_level'], bins=30, kde=True, color='#ff7f0e')  # Orange
plt.title('Average Glucose Level Distribution', fontsize=14)
plt.xlabel('Avg Glucose Level')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Boxplot: BMI vs Stroke
plt.figure(figsize=(10, 6))
sns.boxplot(x='stroke', y='bmi', hue='stroke', data=df, palette={0: '#2ca02c', 1: '#d62728'}, dodge=False)
plt.title('BMI by Stroke Occurrence', fontsize=14)
plt.xlabel('Stroke')
plt.ylabel('BMI')
plt.xticks([0, 1], ['No Stroke', 'Stroke'])
plt.legend([],[], frameon=False)  # Hide double legends
plt.show()
```

```python
# 4. Boxplot: Age vs Stroke
plt.figure(figsize=(10, 6))
sns.boxplot(x='stroke', y='age', hue='stroke', data=df, palette={0: '#9467bd', 1: '#8c564b'}, dodge=False)
plt.title('Age by Stroke Occurrence', fontsize=14)
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.xticks([0, 1], ['No Stroke', 'Stroke'])
plt.legend([],[], frameon=False)  # Hide double legends
plt.show()
```

```python
# 5. Scatter Plot: Age vs Glucose Level colored by Stroke
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='avg_glucose_level', hue='stroke', data=df, palette={0: '#17becf', 1: '#e377c2'})  # Teal and Pink
plt.title('Age vs Glucose Level (Stroke Colored)', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Avg Glucose Level')
plt.legend(title='Stroke', loc='best')
plt.show()
```

```python
# 6. Correlation Heatmap
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```

```python
# 7. Pairplot for numerical features
numerical_cols = ['age', 'avg_glucose_level', 'bmi']
sns.pairplot(df[numerical_cols + ['stroke']], hue='stroke', diag_kind='kde', palette={0: '#2ca02c', 1: '#d62728'})
plt.suptitle('Pairplot of Numerical Features by Stroke Status', y=1.02)
plt.show()
```
