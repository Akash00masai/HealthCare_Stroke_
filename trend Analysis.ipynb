```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# load data
df = pd. read_csv('cleaned_healthcare_dataset_stroke.csv')
```

```python
# Create Age Groups
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
# Trend Analysis for Stroke Occurrences

# 1. Stroke Rate vs Age
age_stroke_rate = df.groupby('age')['stroke'].mean()

plt.figure(figsize=(12,6))
sns.lineplot(x=age_stroke_rate.index, y=age_stroke_rate.values, color='red')
plt.title('Stroke Rate by Age', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Stroke Rate')
plt.grid(True)
plt.show()
```

```python
# 2. Stroke Rate by BMI
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='stroke', y='bmi', palette='pastel')
plt.title('BMI Distribution for Stroke vs Non-Stroke', fontsize=16)
plt.xlabel('Stroke')
plt.ylabel('BMI')
plt.show()
```

```python
# 3. Stroke Rate by Average Glucose Level
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='stroke', y='avg_glucose_level', palette='coolwarm')
plt.title('Glucose Level Distribution for Stroke vs Non-Stroke', fontsize=16)
plt.xlabel('Stroke')
plt.ylabel('Average Glucose Level')
plt.show()
```

```python
# 4. Stroke vs Smoking Status
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='smoking_status', hue='stroke', palette='muted')
plt.title('Stroke by Smoking Status', fontsize=16)
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()
```

```python
# 5. Stroke vs Hypertension
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='hypertension', hue='stroke', palette='Set2')
plt.title('Stroke by Hypertension Status', fontsize=16)
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()
```

```python

# 6. Stroke vs Heart Disease
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='heart_disease', hue='stroke', palette='Set1')
plt.title('Stroke by Heart Disease Status', fontsize=16)
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()
```
